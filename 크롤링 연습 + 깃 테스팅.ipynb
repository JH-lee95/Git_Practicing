```python
group_nav=soup.find("div",class_='group_nav')
print(group_nav)
print(type(group_nav))
```

```python
list_nav=group_nav.find_all('ul',class_="list_nav")  #ul 태그에서, list_nav 클래스를 찾는다
list_nav

```

```python
result_list=[]
for nav in list_nav[:2]:
    anchor_tag=nav.find_all('a')
    for a in anchor_tag:
        title=a.text
        url=a.get('href')  #href 속성에 접근

        result_list.append({"Title" : title,
                      "Url": url})
            
```

#### 방탄소년단 뉴스 리스트 크롤링

```python
bts=requests.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8")
```

```python
bts.text
```

```python
soup_bts=bs(bts.text)
```

```python
titles=soup_bts.select("ul.type01>li>dl>dt>a")
```

```python
#css selector, 
descs=soup_bts.select("ul.type01>li>dl>dd:nth-child(3)") ## ":nth-child()" , n번째 자식 가져오기
```

```python
article_list=[]
for a_tag,desc_tag in zip(titles,descs):
    article_list.append({
        "title":a_tag.text,
        "url":a_tag.get("href"),
        "desc":desc_tag.text
        
    })
    
```

```python
article_list
```

#### find로 찾기

```python
soup1=bs(bts.text)
ul_tags=soup1.find('ul',class_='type01')
news_li=ul_tags.find_all('li',recursive=False) #Recursive=False 가 되면 자식의 "li"는 가져오지 않음
news_li
```

```python
result_list=[]
for news in news_li:
    title_tag = news.find("a",class_="_sp_each_url")
    title=title_tag.text
    url=title_tag.get('href')
    desc=news.find_all("dd")
    desc=desc[1].text
    result_list.append({'title':title,
                       'url':url,
                        "desc":desc
                       })
```

```python
result_list
```

### 네이버 블로그 크롤링 모듈

```python
def search_blog(url,where,query,page):
    params={
        'where':where,
        'query':query,
        'start':page,
    }
    resp=requests.get(url,params)
    soup=bs(resp.text)
    blog_list=soup.select("#elThumbnailResultArea>li")
    
    page_list=[]
    for tag in blog_list:
        title = tag.find("a",class_="sh_blog_title")
        url = title.get("href")
        content=tag.find("dd",class_="sh_blog_passage")

        desc={
            "Title":title.text,
            "URL":url,
            "Content":content.text
        }

        page_list.append(desc)
    
    return page_list,"page number : ",page
```

```python
search_blog("https://search.naver.com/search.naver","post","자연어처리",1)
```
