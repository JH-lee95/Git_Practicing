# BeautifulSoup

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(resp.text)
```

```python
type(soup)  #resp.text와 달리 타입이 bs4 객체
```

```python
print(soup.prettify()) #예쁘게 보이기
```

```python
#soup.<tag_name> 태그 가져오기

print(soup.title) #타이틀 태그 가져오기    
print(type(soup.title))
```
