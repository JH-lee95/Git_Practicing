# 오차역전파

 : 경사하강법의 확장 개념

1. 다중 퍼셉트론에서 오차를 구하는 방법

2. 결과값의 오차를 구해 이를 토대로 하나 앞선 가중치를 차례로 거슬러 올라가가며 조정해 감

3. 출력층에서 시작해 입력층으로 거슬러 올라감



 - 오차 역전파 구동 방식의 정리
  1. 임의의 초기 가중치를 준 뒤 결과를 계산한다.
  
  2. 계산 결과와 우리가 원하는 값 사이의 오차를 구하낟.
  
  3. 경사 하강법을 이용해 바로 앞 가중치를 오차가 작아지는 방향으로 업데이트한다.
  
  4. 1~3 과정을 더이상 오차가 줄어들지 않을 때까지 반복한다.
  
  
  
  
  
- 오차가 작아지는 방향으로 업데이트 한다 = 미분값이 0에 가까워지는 방향으로 나아간다는 말.

    가중치에서 기울기를 빼도 값의 변화가 없을 때 까지 계속해서 가중치 수정 작업을 반복하는 것

## 실습

```python
import tensorflow as tf
import numpy as np
```

```python
learning_rate=0.1
tf.set_random_seed(0)
np.random.seed(0)
```

```python
x_data=[[0,0],[0,1],[1,0],[1,1]]
y_data=[[0],[1],[1],[0]]

X=tf.placeholder(tf.float32,[None,2])
Y=tf.placeholder(tf.float32,[None,1])
```

```python
W1=tf.Variable(tf.random_normal([2,2]),name="weight1")
b1=tf.Variable(tf.random_normal([2]),name="bias1")
layer1=tf.sigmoid(tf.matmul(X,W1)+b1)
W2=tf.Variable(tf.random_normal([2,1]),name="weight2")
b2=tf.Variable(tf.random_normal([1]),name="bias2")
hypothesis=tf.sigmoid(tf.matmul(layer1,W2)+b2)
```

```python
cost=-tf.reduce_mean(Y*tf.log(hypothesis)+(1-Y)*tf.log(1-hypothesis))
train=tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

predicted=tf.cast(hypothesis>0.5,dtype=tf.float32)
accuracy=tf.reduce_mean(tf.cast(tf.equal(predicted,Y),dtype=tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(10001):
        sess.run(train,feed_dict={X:x_data,Y:y_data})
        if step%100==0:
            print(step,sess.run(cost,feed_dict={X:x_data,Y:y_data}),sess.run(W2))
        
    h,c,a=sess.run([hypothesis,predicted,accuracy],feed_dict={X:x_data,Y:y_data})
    print("\nHypothesis : \n",h,"\nCorrect : \n", "\nAccuracy : \n",a)
```
