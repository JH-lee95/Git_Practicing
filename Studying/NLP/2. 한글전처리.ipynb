```python
from konlpy.tag import *
```

## 토큰화

```python
text="인생은 모두가 함께하는 여행이다. 매일매일 사는 동안 우리가 할 수 있는 건 최선을 다해 이 멋진 여행을 만끽하는 것이다.."
print(text.split(" "))
```

```python
#코모란
from konlpy.tag import Komoran
komoran =Komoran()

#morphs : 형태소 분석
komoran_tokens=komoran.morphs(text)
print(komoran_tokens)
```

```python
#한나눔
from konlpy.tag import Hannanum
hannanum=Hannanum()
hannanum_tokens=hannanum.morphs(text)
print(hannanum_tokens)
```

```python
# Okt(Twitter Tokenizer)
from konlpy.tag import Okt
okt=Okt()
okt_tokens=okt.morphs(text)
print(okt_tokens)
```

```python
# Kkma

from konlpy.tag import Kkma
kkma=Kkma()
kkma_tokens=kkma.morphs(text)
print(kkma_tokens)
```

## 품사 부착

```python
#코모란
komoranTag=[]
for token in komoran_tokens:
    komoranTag += komoran.pos(token)
print(komoranTag)
```

```python
#한나눔
hannanumTag=[]
for token in hannanum_tokens:
    hannanumTag += hannanum.pos(token)
print(hannanumTag)
```

```python
#Okt

oktTag=[]
for token in okt_tokens:
    oktTag+=okt.pos(token)
print(oktTag)
```

```python
#Kkma

kkmaTag=[]
for token in kkma_tokens:
    kkmaTag += kkma.pos(token)
print(kkmaTag)
```

## 불용어처리

```python
from collections import Counter
Counter(oktTag).most_common() #가장 빈도수가 높은 단어순으로 정렬
```

```python
#불용어처리
stopPos=["DeterMiner","Adverb","Conjuction","Josa","PreEomi","Eomi","Suffix","Punctuation","Foreign","Alpha","Number","Unknown"]
stopWord=["을","은","가","할","수","건"]
word=[]

for tag in oktTag:
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
            word.append(tag[0])

print(word)
```

## 기사 크롤링 후 태깅

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.yna.co.kr/view/AKR20200204081000017?section=it/it"
resp=requests.get(url)
```

```python
resp
```

```python
soup=BeautifulSoup(resp.text,"lxml")
```

```python
news_tag=soup.select("#articleWrap > div.content01.scroll-article-zone01 > div > div > div.story-news.article")
```

```python
lines=news_tag[0].find_all("p")
```

```python
article=""

for line in lines:
    article += line.text    
```

```python
article
```

```python
article=article.replace("(서울=연합뉴스) 홍지인 기자 = 구글의 칼라 브롬버그 \'공익을 위한 AI\' 프로그램 리드가 4일 역삼동 구글코리아에서 화상 기자간담회를 갖고 있다. 2020. 2. 4. (서울=연합뉴스) 홍지인 기자 = ","")
```

```python
article=article.replace(" ljungberg@yna.co.kr<저작권자(c) 연합뉴스, 무단 전재-재배포 금지>2020/02/04 11:33 송고","")
```

```python
article
```

```python
from konlpy.tag import Mecab
```

```python
mecab=Mecab()
```

```python
mecab_tokens=mecab.morphs(article)
```

```python
mecab_tokens
```

```python
mecab_pos=[]
for i in mecab_tokens:
    mecab_pos +=mecab.pos(i)
```

```python
mecab_pos
```

```python
Counter(mecab_pos).most_common()
```

```python
# 불용어 처리
stopwords=["은","는","(",")","을","를","보","다","더"]

words=[]

for word in mecab_pos:
    if word[0] not in stopwords:
        words.append(word[0])
```

```python
print(words)
```
