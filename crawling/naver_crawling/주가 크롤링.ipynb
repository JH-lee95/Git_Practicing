## 주가 크롤링 함수화

```python
def get_day_price(code,page_num):
    
    result_stock=[]
    for page in range(1,page_num+1):  #검색하고자 하는 페이지 수 만큼 반복
        base_url="https://finance.naver.com//item/sise_day.nhn"   #해당 iframe의 url
        params={
            "code":code,
            "page":page}   #url에 들어갈 파라미터 설정 , 'code' : 종목코드 , "page" : 페이지 번호
        
        
        
        resp_stock=requests.get(base_url,params)   #해당 url에 리퀘스트
        soup_stock=bs(resp_stock.text) #Beautifulsoup 으로 받아온 url parsing 
        
        tr_tags=soup_stock.select("table.type2 tr") #크롤링 하고자 하는 태그에 접근
        

        th_tags=tr_tags[0].find_all("th")  #헤더만 검색, 딕셔너리의 key가 될 요소들
        title_list=[]
        for th in th_tags:
            title_list.append(th.text)

        for tr_tag in tr_tags:   #tr 태그를 돌면서
            td_tags=tr_tag.find_all('td') # 각 tr태그 내에 있는 td 태그들 검색
            if len(td_tags)<7: #td 태그들의 길이가 7이하면 무시 
                continue
            data_list=[]
            for td in td_tags: #td 태그내 각 정보들에 접근
                data_list.append(td.text.strip()) #해당 태그의 텍스트만 (정보만)
            result_stock.append(dict(zip(title_list,data_list)))
            
    return result_stock
```

```python
sise=get_day_price("005930",3)
sise
```
