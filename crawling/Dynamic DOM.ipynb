## 업비트 차트

```python
base_url="https://crix-api-cdn.upbit.com/v1/crix/candles/months"  #캔들 차트 url, (리퀘스트를 계속 보내는 url 가져옴)
params={
    'code':"CRIX.UPBIT.KRW-BTC",
    'count':200
}
resp=requests.get(base_url,params=params)
resp
```

```python
import json
```

```python
month_candle=json.loads(resp.text) #파이썬 객체화 (리스트 안의 딕셔너리)
month_candle
```

```python
type(month_candle[0])
```

```python
month_candle[0]['lowPrice']
```

```python
month_json=json.dumps(month_candle) #json 파일화 시키기
month_json
```

```python
month2=resp.json() #requests 의 json 함수
month2
```

## 이미지 저장

```python
url="https://search.pstatic.net/common/?src=https%3A%2F%2Fimgnews.pstatic.net%2Fimage%2Forigin%2F001%2F2020%2F07%2F31%2F11783389.jpg&amp;type=ofullfill80_80_q75_re2"
resp=requests.get(url)
resp
```

```python
resp.content #이미지를 바이너리로 저장
```

```python
with open('temp.jpg',"wb") as f:  #바이너리를 파일화 
    f.write(resp.content)
```
